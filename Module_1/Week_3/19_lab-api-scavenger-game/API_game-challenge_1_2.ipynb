## 1. Obtain the full list of forks created from the main lab repo via Github API.

```python
response = requests.get('https://api.github.com/repos/ironhack-datalabs/databer0619/forks', auth=('TheManohar', '9232ab66b9c233b2589a87c7c28822699ab70482'));
```

```python
data = response.json()
```

```python
from pandas.io.json import json_normalize
```

```python
forks = json_normalize(data);
forks.head(10)
```

## 2. Loop the JSON response to find out the language attribute of each fork. Use an array to store the language attributes of each fork.

```python
languages = []

for language in forks['language']:
    if language not in languages:
        languages.append(language)
        
print(languages)
```

```python
languages = list(forks.language.unique());
print(languages);
```
